### a)
The algorithm `ALG1` first **sorts** the array `A` and then **iterates** through it to find a discrepancy between the array indices and the values they contain. The core idea is that if no number were missing, a sorted array `A` (using 1-based indexing) of numbers from `{0, ..., n}` would have the property that `A[i] = i-1` for all `i` from `1` to `n`. The missing number `x` creates a "shift" in this pattern due to the offset of indices and array values.

To prove the algorithm is correct, we must show it works for all possible values of the missing number `x`. We can split this into two cases.

---

#### Case 1: The missing number `x` is less than `n` ($0 \le x < n$)

1.  **Array State after Sorting**:
    After sorting, the array `A` will contain all numbers from `{0, ..., n}` except `x`, in ascending order.
    The contents of `A` will be:
    $$A = [0, 1, 2, \dots, x-1, x+1, \dots, n]$$
    Using 1-based indexing as in the pseudocode, we have:
    - For indices `i` from `1` to `x`:  `A[i] = i-1` as we stated.
    - At index `i = x+1`: The value is `A[x+1] = x+1`,  because the missing number disrupts the pattern of `A[i] = i-1`.

2.  **Loop Execution**:
    The algorithm iterates from `i=1` to `n`.
    -   **For `i` from `1` to `x`**: The condition `if A[i] == i` is checked. Since `A[i] = i-1` in this range, the check becomes `if i-1 == i`, which is **false** and the loop continues.
    -   **At `i = x+1`**: The condition `if A[i] == i` is checked. At this index, `A[x+1] = x+1`. The check becomes `if x+1 == x+1`, which is **true** because of the missing number.

3.  **Return Value**:
    The condition is met for the first time at `i = x+1`. The procedure then executes `return i-1`.
    $$\text{return} \ (x+1) - 1 = x$$
    This is the correct missing number.

---

#### Case 2: The missing number `x` is `n`

1.  **Array State after Sorting**:
    If `x=n` is missing, the array `A` contains all numbers from `0` to `n-1`. After sorting, `A` will be:
    $$A = [0, 1, 2, \dots, n-1]$$
    Using 1-based indexing, this means that for every index `i` from `1` to `n`, the property `A[i] = i-1` holds true.

2.  **Loop Execution**:
    The algorithm iterates from `i=1` to `n`. At each step, it checks the condition `if A[i] == i`.
    - Since `A[i] = i-1` for all `i` in the loop, the check is always `if i-1 == i`, which is always **false**.

3.  **Return Value**:
    The `if` condition is never satisfied. The loop completes without returning. The procedure then moves to the final line, which is `return n`. This is the correct missing number, `x=n`.

---

#### Conclusion

Since the algorithm correctly computes the missing number `x` in all possible cases (`x < n` and `x = n`) and it is guaranteed that the array contains exactly one number from 0..n with one missing number the algorithm is **correct**.

### b)
### Proof of Correctness for ALG2

The algorithm `ALG2` finds the difference between an expected sum and an actual sum. The set of numbers *should* be `{0, 1, ..., n}` with 'n' as the length of the array which we know. But the array `A` is missing one number, `x`. The algorithm calculates the sum of all numbers in the incomplete array `A` and subtracts it from the sum of a known, complete sequence of numbers (the sum of the numbers from 0 to n).

To prove the algorithm is correct, we need to proof that the algorithm correctly calculates the difference between the two sums.

---
### 1. The two sums

* Let `S_full` be the sum of all numbers in the complete set `{0, 1, ..., n}`.
    $$S_{full} = \sum_{k=0}^{n} k$$

* Let `S_A` be the sum of all the numbers actually present in the input array `A`. Since `A` contains `n` elements, this sum is:
    $$S_A = \sum_{i=1}^{n} A[i]$$

#### 2. The Key Relationship

The set of numbers in array `A` is the same as the complete set `{0, 1, ..., n}` but with the number `x` removed. Therefore, the sum of `A`'s elements (`S_A`) is simply the sum of the full set (`S_full`) minus the missing number `x`.

$$S_A = S_{full} - x$$

Substituting the definition of `S_full`, we get:

$$S_A = \left(\sum_{k=0}^{n} k\right) - x$$

---

#### 3. Analyzing the Algorithm's Calculation

The algorithm initializes a variable `y` to 0 and then iterates from `i=1` to `n`. In each step, it calculates `y = y + i - A[i]`. Let's analyze the final value of `y` after the loop has finished.

The final value is the sum of all the `i` values minus the sum of all the `A[i]` values:

$$y = \sum_{i=1}^{n} (i - A[i])$$

Using the properties of summation, we can split this into two separate sums:

$$y = \left(\sum_{i=1}^{n} i\right) - \left(\sum_{i=1}^{n} A[i]\right)$$

We recognize the second term as `S_A`. So:

$$y = \left(\sum_{i=1}^{n} i\right) - S_A$$

---

#### 4. Final Derivation

Now, we substitute our expression for `S_A` from step 2 into the equation for `y`:

$$y = \left(\sum_{i=1}^{n} i\right) - \left[ \left(\sum_{k=0}^{n} k\right) - x \right]$$

To simplify, let's expand the sum from `k=0` to `n`:

$$\sum_{k=0}^{n} k = 0 + \sum_{k=1}^{n} k = \sum_{i=1}^{n} i$$

(Note that the index variable `k` or `i` doesn't matter, the sum is the same).

Now substitute this back into the equation for `y`:

$$y = \left(\sum_{i=1}^{n} i\right) - \left[ \left(\sum_{i=1}^{n} i\right) - x \right]$$

Distributing the negative sign gives:

$$y = \left(\sum_{i=1}^{n} i\right) - \left(\sum_{i=1}^{n} i\right) + x$$

The two summations cancel each other out, leaving:

$$y = x$$

---

#### Conclusion

The final value of `y` returned by the algorithm is exactly equal to the missing number `x`. Therefore, the algorithm is **correct**.

First we can simplify the functions, by using Big O-Notation, since we can ignore constant factors for large n.

-   $2n^2 - 4n \rightarrow O(n^2)$
-   $n^2 \log n \rightarrow O(n^2 \log n)$
-   $n \log n \rightarrow O(n \log n)$
-   $2^n \rightarrow O(2^n)$
-   $\log n \rightarrow O(\log n)$
-   $n \log(\log n) \rightarrow O(n \log(\log n))$

Considering the law of transitivity from exercise 3b, the final ordering from slowest to fastest asymptotic growth is:

$$\log n < n \log \log n < n \log n < 2n^2 - 4n < n^2 \log n < 2^n$$

Proofs by considering the mathematical asymptotic behavior:

**Proof for $\log n \in o(n \log \log n)$**

$$\lim_{n \to \infty} \frac{\log n}{n \log \log n} = \lim_{n \to \infty} \frac{\log n}{n} \cdot \frac{1}{\log \log n} = 0$$

**Proof for $$\lim_{n \to \infty} \frac{\log n}{n}$$**

$$= \lim_{n \to \infty} \frac{1/n}{1} \quad (L'Hôpital)$$

$$= \lim_{n \to \infty} \frac{1}{n} = 0$$
(This holds because both $\frac{\log n}{n}$ and $\frac{1}{\log \log n}$ move towards 0 as $n \to \infty$)

---

**Proof for $n \log \log n \in o(n \log n)$**

$$\lim_{n \to \infty} \frac{n \log \log n}{n \log n} = \lim_{n \to \infty} \frac{\log \log n}{\log n}$$

Using L'Hôpital's Rule:

$$= \lim_{n \to \infty} \frac{\frac{d}{dn}(\log \log n)}{\frac{d}{dn}(\log n)} = \lim_{n \to \infty} \frac{\frac{1}{n \log n}}{\frac{1}{n}} = \lim_{n \to \infty} \frac{n}{n \log n} = \lim_{n \to \infty} \frac{1}{\log n} = 0$$

---

**Proof for $n \log n \in o(2n^2 - 4n)$**

$$\lim_{n \to \infty} \frac{n \log n}{2n^2 - 4n} = \lim_{n \to \infty} \frac{n \log n}{n \cdot (2n - 4)} = \lim_{n \to \infty} \frac{\log n}{2n - 4}$$

Using L'Hôpital's Rule:

$$= \lim_{n \to \infty} \frac{1/n}{2} = 0$$

---

**Proof for $2n^2 - 4n \in o(n^2 \log n)$**

$$\lim_{n \to \infty} \frac{2n^2 - 4n}{n^2 \log n} = \lim_{n \to \infty} \frac{n \cdot (2n - 4)}{n^2 \log n} = \lim_{n \to \infty} \frac{2n - 4}{n \log n}$$

Using L'Hôpital's Rule:

$$= \lim_{n \to \infty} \frac{2}{\log n + 1} = 0$$

---

#### Proof for $n^2 \log n \in o(2^n)$

$$\lim_{n \to \infty} \frac{n^2 \log n}{2^n}$$

$$= \lim_{n \to \infty} \frac{2n \log n + n}{2^n \ln 2} \quad (L'Hôpital)$$

$$= \lim_{n \to \infty} \frac{2 \log n + 3}{2^n (\ln 2)^2} \quad (L'Hôpital)$$

$$= \lim_{n \to \infty} \frac{2/n}{2^n (\ln 2)^3} \quad (L'Hôpital)$$

$$= \lim_{n \to \infty} \frac{2}{n \cdot 2^n (\ln 2)^3} = 0$$
